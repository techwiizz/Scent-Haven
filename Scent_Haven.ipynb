```python
dfc = df['Price'].corr(df['Quantity'])
print(f"Correlation between Price and Quantity: {dfc}")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['DayOfWeek'] = df['Date'].dt.day_name()
df['Weekend'] = df['DayOfWeek'].isin(['Saturday', 'Sunday'])
sales_by_day = df.groupby('Weekend')['Total '].sum().reset_index()
print(sales_by_day)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['DayOfWeek'] = df['Date'].dt.day_name()
average_sales_by_day = df.groupby('DayOfWeek')['Total '].mean().reset_index()
day_with_highest_sales = average_sales_by_day.loc[average_sales_by_day['Total '].idxmax()]

print("Day with the highest average sales:")
print(day_with_highest_sales)
```
